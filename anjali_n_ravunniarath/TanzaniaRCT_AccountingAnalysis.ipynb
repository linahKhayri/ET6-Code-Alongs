```python
import pandas as pd

# Load the Excel file and the specific sheet
file_path = '/Users/anjalinatarajan/Documents/Example File/Example_AccountingData-TanzaniaRCT.xlsx'
sheet_name = 'Data - for Calc'

# Read the Excel file
dataframe = pd.read_excel(file_path, sheet_name=sheet_name, skiprows=3)
```

```python
dataframe.head(10)
```

```python
# Relevant categories
category_columns = ['Deliver coupons', 'Refund Dukas', 'Administration', 'Program operations', 'Management', 'Fundraising', 'Public cloud platform', 'Travel and local transport']

# Granular divisions for each category
granular_categories = {
    'Deliver coupons': ['Printing_Designing', 'household_visits', 'Delivering_coupons'],
    'Refund Dukas': ['recruitment', 'visits_payments', 'Collecting_coupons', 'Obtaining_coupons', 'Reviewing_redeemed_coupons', 
                     'Distributing_incentives'],
    'Program operations': ['laptops_Airtime', 'Mystery_shopping', 'Solving_issues', 'Monitoring_progress', 
                           'Program_Administrative_tasks', 'Monitoring_field_progress'],
    'Travel and local transport': ['Meeting', 'Travel_to_field']
}
```

```python
category_sums = {}

for category in category_columns:
    category_sum = dataframe.loc[dataframe[category] == 1, 'Amount (USD)'].sum()
    category_sums[category] = category_sum
```

```python
category_sums
```

```python
# exclusion columns
exclude_cols = ['Evaluation-induced', 'Outside Analytical Time Window','Activity unrelated to TZ-RCT']

#calculating costs for categories with exclusions
for category in category_columns:
    filtered_df = dataframe[
        (dataframe['Evaluation-induced'] != 1) &
        (dataframe['Outside Analytical Time Window'] != 1) &
        (dataframe['Activity unrelated to TZ-RCT'] != 1) &
        (dataframe[category] == 1)
    ]
    category_sum_w_exclusions = filtered_df['Amount (USD)'].sum()
    category_sums_w_exclusions[category] = category_sum_w_exclusions
```
